```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# All possible outcomes when tossing a coin
outcomes = ["H", "T"]

def coin_flipper():
    trails_list= [100, 500, 1000, 2000, 3000, 4000, 5000, 10000]
    nums_flips= 500
    
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 12))
    axes = axes.flatten()
    for idx, trails in enumerate(trails_list):
        obs_list= []                               # creating an empty list to store the summation of n toss
        for _ in range(trails):
            res= np.random.choice(outcomes, nums_flips)
            def convert(res):                      # Converting H(Head) to '+1' and T(Tail) to '-1'
                converted_list= []
                for outcome in res:
                    if outcome=='H':
                        converted_list.append(1)
                    else:
                        converted_list.append(-1)
                return converted_list    
                                                   
            converted_res= convert(res)            # Converting and calculating the sum of the outcomes for each trial
            obs_list.append(np.sum(converted_res)) # Appending the sum of n toss to the list   
        
        sns.histplot(data=obs_list, bins=20, kde=True, ax=axes[idx])
        axes[idx].set_title(f'For {trails_list[idx]} Observations:\n', fontsize=15)
        axes[idx].set_xlabel('Sum of 20 Flips')
        axes[idx].set_ylabel('Density')

    plt.tight_layout()
    plt.show()
coin_flipper()
```

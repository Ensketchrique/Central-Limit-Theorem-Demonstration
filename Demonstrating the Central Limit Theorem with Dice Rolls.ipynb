```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# All possible outcomes when we roll a die
outcomes= [num for num in range(1,7)]
# Program to perform the experiments: Recording n number of obeservations 
# for the sum of all outcomes when a die is rolled m number of times.
def dice_roller():
    obs_lst= [100, 500, 1000, 2000, 5000, 10000]            # Different observation sizes     
    num_rolls= 10                                           # Number of times we roll the die in each observation
    series_list= []                                         # To store each series for plotting later
    for i in range(len(obs_lst)):
        obs_sum = []                                                # Stores the sum for each observation size
        for j in range(obs_lst[i]):    
            rolls_sum= sum(np.random.choice(outcomes, num_rolls))   # Rolling the die 10 times and adding the outcomes
            obs_sum.append(rolls_sum)                               # Appending the sum of 10 outcomes
        
        series_list.append(pd.Series(obs_sum))                      # Storing the series for each experiment
    fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))    # Adjust the size as needed
    axes = axes.flatten()                                           # Flatten the axes array for easy iteration 
    for i, col in enumerate(series_list):
        sns.histplot(data=col, bins=20, kde=True, ax= axes[i])      # ploting histograms in subplots 
        axes[i].set_title(f'For {obs_lst[i]} Observations:\n', fontsize=15)  # Set title for each plot
        axes[i].set_xlabel("Sum of 10 Rolls")  # X-axis label
        axes[i].set_ylabel('Density')  # Y-axis label
        plt.subplots_adjust(hspace=.5, wspace=.2)
        
dice_roller()
```
